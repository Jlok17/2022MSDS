```python
#Building Blocks
year = '2019'
url_link = "https://www.pro-football-reference.com/years/{}/rushing.htm"

#Combining the URL + year Strings together
url = url_link.format(year)
url
```

```python
years = [2014,2015,2016,2017,2018,2019,2020,2021]
url_link = 'https://www.pro-football-reference.com/years/{}/rushing.htm'

for year in years:
  url = url_link.format(year)
  print(url)
```
